```python
def TotalGoals (name):
    goals = 0
    if len(name) == 0:
        return "no player"
    else:
        for n in range(0,len(newIndex["Player"])):
            if newIndex["Player"][n] == name:
                goals = goals + newIndex["G"][n]
        goalsstr = str(goals)
        return name + " scored " + goalsstr + " goals."
```
